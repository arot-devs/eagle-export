```python
import pandas as pd

# Define conditions in a list: (column_name, lambda condition, tag_to_add)
conditions = [
    ("focus_measure", lambda x: x < 100, "blurry"),
    ("noise_level", lambda x: x > 70, "noisy"),
    ("noise_level", lambda x: x <10, "un-noisy"),
]

# Apply conditions efficiently
def add_tags(row):
    tags = row["tags"]
    if not isinstance(tags, list):  # Ensure it's a list
        tags = [] if pd.isna(tags) else [tags]

    for col, condition, tag in conditions:
        if condition(row[col]) and tag not in tags:
            tags.append(tag)

    return tags

df["tags"] = df.apply(add_tags, axis=1)
```

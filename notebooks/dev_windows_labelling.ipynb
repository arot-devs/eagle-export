```python
from eagle_exporter.core import build_dataframe, write_df_to_json

# Build DataFrame without loading images
folder = r"E:\Datasets\eagle_quick_rate_novelai\eagle_template.library"
df = build_dataframe(folder, include_images=False, keep_paths=True)
df
```

```python
from procslib import get_model_keys, get_model

# model = get_model("custom_image", model_path="distill-lab/datatmp_nai-distill_01_combined_eagle.library_classification_batch1234",
#                   task="classification")


model = get_model("custom_image", model_path="incantor/datatmp_nai-distill_01_combined_eagle.library_classification_batch123",
                  task="classification")

res = model.infer_many(df["image_path"].tolist())
```

```python
res
```

```python
# label = star_0, confidence > 0.9: value = 0
# label = star_0, confidence < 0.9: value = 1
# label = star_1, confidence > 0.9: value = 2
# label = star_1, confidence < 0.9: value = 3

def assign_value(label, confidence):
    if label == "star_0" and confidence > 0.9:
        return 0
    elif label == "star_0" and confidence < 0.9:
        return 1
    elif label == "star_1" and confidence > 0.9:
        return 2
    elif label == "star_1" and confidence < 0.9:
        return 3
    else:
        return 6

res["image_path"] = res["path"]
res["value"] = res.apply(lambda x: assign_value(x["label"], x["confidence"]), axis=1)
res["tags"] = res["value"].apply(lambda x: [f"pred_{x}"])

res
```

```python
# df = df.drop(columns=["tags_x", "tags_y"])
df = df.drop(columns=["tags"])
```

```python
df = df.merge(res[["image_path", "tags"]], on="image_path", how="left")
```

```python
write_df_to_json(df)
```
